## Part 3: Data Transformations

```python
WHO_df = pd.read_csv('C:/Users/rgs88/Documents/Principles of Data Science/TB_burden_countries_2014-09-29.csv')
WHO_df.dtypes
```

```python
for col in WHO_df:
        if WHO_df[col].dtype == 'float64':
            WHO_df[col].fillna(WHO_df[col].mean(), inplace = True)
        else:
            WHO_df[col].fillna(0, inplace = True)
```

```python
WHO_df
```

```python
plt.hist(WHO_df['e_prev_100k_hi'], bins=20)
```

```python
WHO_df2 = WHO_df.select_dtypes(exclude=['object'])
column_means = WHO_df2.mean()
WHO_df2 = WHO_df2.fillna(column_means)
WHO_df_log = np.log(WHO_df2)
```

```python
normalisedlog_df = (WHO_df_log - WHO_df_log.min())/(WHO_df_log.max()-WHO_df_log.min())
```

```python
normalisedlog_df
```
